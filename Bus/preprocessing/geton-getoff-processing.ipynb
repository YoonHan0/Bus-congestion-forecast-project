```python
# 승차인원, 하차인원으로 나눠서 각각 엑셀파일로 저장
df_GETON = df.groupby(['YMD_ID', 'HH_ID', 'DAY_CLS', 'WKDAY_HLDAY_CLS', 'BUS_ROUTE_ID'])['GETON_CNT'].sum()#.sort_values(ascending=False)
df_GETOFF = df.groupby(['YMD_ID', 'HH_ID', 'DAY_CLS', 'WKDAY_HLDAY_CLS', 'BUS_ROUTE_ID'])['GETOFF_CNT'].sum()#.sort_values(ascending=False)

df_GETON.to_csv("df_GETON_test.csv", index = True)
df_GETOFF.to_csv("df_GETOFF_test.csv", index = True)
```
